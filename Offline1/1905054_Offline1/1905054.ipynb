# Data cleaning

```python
# handling the null values with the mean of the column 
dataframe.fillna(dataframe.select_dtypes(include='number').mean(), inplace=True)
#dataframe.isnull().sum()
```

```python
#Droping duplicates, keeping only one copy of data each
dataframe.drop_duplicates(inplace=True)
```

```python
#Drop the rows when target is missing

dataframe.dropna(subset=['Attrition'], inplace=True)
dataframe.isnull().sum()
```

# Creation of input and output features:


```python
#creating input and output data

features = dataframe.drop('Attrition', axis=1)
labels = dataframe['Attrition']
#input
features
```

```python
#output

labels
```

# Conversion of features into numeric values:

```python
#converting the columns into numeric features
from sklearn.preprocessing import LabelEncoder # type: ignore

#label encoding the target
encoder = LabelEncoder()
labels = dataframe['Attrition']
labels = encoder.fit_transform(labels)
labels
```

```python
# one-hot encoding the features
features = pd.get_dummies(features)
features
```

# Scaling of the features:

```python
#Standard scaling 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_std = scaler.fit_transform(features)
features_std
```

```python
#MinMax scaling 
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
features_minmax = scaler.fit_transform(features)
features_minmax
```

```python
#Using function
from sklearn.preprocessing import StandardScaler, MinMaxScaler

#Defining the scaling function
def scale_features(features, scaler_type='standard'):
    if scaler_type == 'standard':
        scaler = StandardScaler()  
    elif scaler_type == 'minmax':
        scaler = MinMaxScaler()   
    else:
        raise ValueError("Invalid scaler_type. Use 'standard' or 'minmax'.")
    
    # Applying the chosen scaler
    features_scaled = scaler.fit_transform(features)
    
    return features_scaled
```

```python
features_show = scale_features(features,'standard')
features_show
```

# Correlation Analysis:

```python
features_df = pd.DataFrame(features_show, columns=features.columns) #scaled featured dataframe
labels_df = pd.DataFrame(labels, columns=['Attrition'])
```

```python
#Correlation analysis of features with target
labels_series = labels_df['Attrition']
correlation = features_df.corrwith(labels_series)
correlation
```

```python
#Selecting top 20 features with the highest correlation with target
correlation_sorted = correlation.abs().sort_values(ascending=False)

top_20_features = correlation_sorted.head(20)
top_20_features
```

# 1D scatter plot

```python
import matplotlib.pyplot as plt # type: ignore
import numpy as np
```

```python
# Iterating top 20 feature names

top_20_feature_names = top_20_features.index
for feature in top_20_feature_names:
    plt.figure(figsize=(8, 4))
    
    plt.scatter(features_df[feature], labels_series, alpha=0.6)
    
    plt.title(f'Scatter Plot of {feature} vs Attrition')
    plt.xlabel(f'{feature}')
    plt.ylabel('Attrition')
    
    plt.show()
```
